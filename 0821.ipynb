> 방문 길이

- 내가 푼 풀이

```python
def solution(dirs):

    ax = [0,0]

    answer = set()

    for _dirs in dirs:

        if _dirs == 'U' and ax[1]!=5:
            ax[1]+=1
            pos = (ax[0], ax[1]-.5)
        elif _dirs == 'D' and ax[1]!=-5:
            ax[1]+=-1
            pos = (ax[0], ax[1]+.5)
        elif _dirs == 'R' and ax[0]!=5:
            ax[0]+=1
            pos = (ax[0]-.5, ax[1])
        elif _dirs == 'L' and ax[0]!=-5:
            ax[0]+=-1
            pos = (ax[0]+.5, ax[1])
        else: pass

        answer.add(pos)

    return len(answer)
```

- 해설지

```python
def solution(dirs):
    move = {'L': (-1, 0), 'R': (1, 0), 'U': (0, 1), 'D': (0, -1)}
    x = y = 0
    edges = set()

    for d in dirs:
        dx, dy = move[d]
        nx, ny = x + dx, y + dy


        if not (-5 <= nx <= 5 and -5 <= ny <= 5):
            continue


        a = (x, y)
        b = (nx, ny)

        if a < b:
            edge = (a, b)
        else:
            edge = (b, a)

        edges.add(edge)


        x, y = nx, ny

    return len(edges)
```
