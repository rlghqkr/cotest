```python
#리스트 사용
def solution(N, stages):

    total = len(stages)
    fail_rate = []

    for i in range(1, N + 1):

        number = stages.count(i)

        if total == 0:
            fail = 0
        else:
            fail = number / total

        fail_rate.append((i, fail))
        total -= number

    fail_rate.sort(key=lambda x: (-x[1], x[0]))

    answer = [stage for stage, rate in fail_rate]
    return answer

#딕셔너리 사용
def solution(N,stages):

    total = len(stages)
    fail_dict = {}

    for i in range(1,N+1):

        number = stages.count(i)

        if total >0:
            fail_dict[i] = number/total
        else:
            fail_dict[i]=0
        total -= number

    fail_list = sorted(fail_dict.items(), key = lambda x : (-x[1],x[0]))
    answer = [fail_info[0] for fail_info in fail_list]

    return answer
```
