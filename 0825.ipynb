```python
n,m = map(int,input("가로 세로 크기 입력").split(" "))
```

```python
graph = []

for i in range(n):
    t = input("n,m지도에 대한{0}번째가로줄".format(i+1)) # "10111"
    t = list(map(int, [i for i in t]))
    # "10111" --> ['1', '0', '1', '1', '1'] --> [1,0,1,1,1]
    graph.append( t)
    
graph
```

```python
# ++ 인접 판정 :  LRUD 이동에 대한 구현을 이용!!!
# ===> L, R, U, D 순서나 방향성은 중요하지 않음!!!
#      이 4가지 방향으로 점검!!
# ===> 방향없이 4가지만 빠지지 않고 하면 되니,,,list 로 세팅!!!
dx = [0,0,-1,1]
dy = [-1,1,0,0]
# --> 내가 4가지 이동에 대한 규칙만 세팅하면 끝!!!! 좌표화!!-->
# 지도에 대한 리스트 구현
# m[가로][세로] ==> 미로 상 한 칸을 접근!!!
```

```python
dx = [0,0,-1,1]
dy = [-1,1,0,0]
```

```python
from collections import deque
```

```python

def bfs_miro( start): # s = (0,0)
    q = deque() # 한 일 세팅
    x_pos,y_pos = start
    q.append( (x_pos,y_pos) )

    while q: #  [ (0,0) ]
        now_x, now_y = q.popleft() # (0,0) --> now_x= 0, now_y =0

        for i in range(4): # 4가지 인접 이동 후보들 점검 LRUD--> len(dx)
            next_x = now_x + dx[i]
            next_y = now_y + dy[i]

            if 0<= next_x < n and 0 <= next_y < m: # 파이썬 리스트 지도--> 좌표-정수: 시작점 0~~
                 
                 if graph[next_x][next_y] == 1:
                    graph[next_x][next_y] = graph[now_x][now_y] + 1
                    q.append( (next_x, next_y))
```
